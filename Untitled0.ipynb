```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file
# Replace 'your_data.csv' with the actual path to your file
# Ensure the file has columns named 'Compound' and 'Energy'
data = pd.read_csv(r"/content/drive/MyDrive/Wild Vegetable Project Nishad/Avishkar/Avishkarepaired.csv")

# Filter the data within the desired energy range
filtered_data = data[(data['Binding Affinity'] <= -8.50) & (data['Binding Affinity'] >= -9.50)]

# Sort by energy for better visualization (optional)
filtered_data = filtered_data.sort_values(by="Binding Affinity")

top_15_data = filtered_data.head(15)

top_15_data = top_15_data.sort_values(by="Binding Affinity")

# Create the scatter plot
plt.figure(figsize=(7, 5))

plt.scatter(filtered_data['Ligand'], filtered_data['Binding Affinity'], color='black', s=60)

# Customize the plot
plt.title("Affinity Energy Scatter Plot", fontsize=12)
plt.xlabel("Compounds", fontsize=10)
plt.ylabel("Affinity Energy (kcal/mol)", fontsize=10)

# Reverse the Y-axis to show negative energies downward
plt.gca().invert_yaxis()

# Rotate X-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=10)

# Add a horizontal reference line at -6 kcal/mol (optional)
plt.axhline(y=-8.50, color='red', linestyle='--', linewidth=1)

plt.grid(True, which='both', linestyle='-', linewidth=0.5, alpha=1.0)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
