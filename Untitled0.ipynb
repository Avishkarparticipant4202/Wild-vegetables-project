```python
import matplotlib.pyplot as plt

# Manually enter your data
observations = list(range(0, 9))  # 9 observations (x-axis)
rmsd_higher = [0.0, 1.0, 1.2, 0.9, 1.3, 1.5, 1.4, 1.6, 1.8]  # Replace with your values
rmsd_lower = [0.0, 0.6, 0.7, 0.5, 0.8, 0.9, 1.0, 1.1, 1.2]   # Replace with your values

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(observations, rmsd_higher, label='RMSD Higher Bound', marker='o', color='red')
plt.plot(observations, rmsd_lower, label='RMSD Lower Bound', marker='o', color='blue')

# Annotate the points with their values
for i, (x, y) in enumerate(zip(observations, rmsd_upper)):
    plt.text(x, y + 0.1, f'{y:.1f}', ha='center', fontsize=9, color='blue')
for i, (x, y) in enumerate(zip(observations, rmsd_lower)):
    plt.text(x, y - 0.1, f'{y:.1f}', ha='center', fontsize=9, color='red')


# Adding labels, title, and legend
plt.xlabel('Observation Number')
plt.ylabel('RMSD Value')
plt.title('RMSD Higher and Lower Bound Plot')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
